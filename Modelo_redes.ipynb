```python
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.models import Sequential
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import numpy as np
```

# **Redes Neuronales**

```python
datos = pd.read_excel('CSV/Datos_finalesV3.xlsx')
datos.head()
```

```python
X = datos[['P1.2', 'P1.3','P1.4','P1.5','P1.6','P1.7','P1.8','P2','P3','P5.1','P5.2','P5.3','P5.4','Tipo comentario', 'sentimiento']].values
y = datos['P1.1'].values

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
y_train = y_train -1
y_train =to_categorical(y_train) 
y_test2 = y_test
y_test= y_test-1
y_test =to_categorical(y_test)
```

```python
model = Sequential([
    Dense(units = 192, input_shape=[15,]),
    Dropout(0.3),
    Dense(units=160, activation='sigmoid'),
    Dropout(0.2),
    Dense(units=64, activation='sigmoid'), 
    Dropout(0.2),
    Dense(units=5, activation='softmax')
])
```

```python
historico = model.compile(optimizer=Adam(0.001),loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=50, batch_size=32, verbose=True)
```

```python
y_pred = np.argmax(model.predict(X_test),axis=1)+1
print(classification_report(y_test2, y_pred=y_pred))
```
